1. Take two 3D coordinates from the user. Find out the distance between these points
using following distance functions.

x and y are vectors representing the coordinates of two points.
Here, k=3

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def euclidean(x,y):
    a=(x-y)**2
    b=np.sum(a)
    c=np.sqrt(b)
    return c
def manhattan(x,y):
    d=np.abs(x-y)
    e=np.sum(d)
    return e
def minkowski(x,y,q):
    f=np.power(np.abs(x-y),q)
    g=np.sum(f)
    h=np.power(g,1/q)
    return h
x=np.array([1,2,3])
y=np.array([4,5,6])
q=float(input())
print("Euclidean Distance: ",euclidean(x,y))
print("Manhattan Distance: ",manhattan(x,y))
print("Minkowski Distance: ",minkowski(x,y,q))
```

2. Import the Iris dataset. Write a program to obtain the Euclidian Distance Matrix for
all the data samples in the feature space. Distance metric is a 2D array, where the
(i,j)th entry represents the distance between the ith and jth sample points in the feature
space.

```python
from sklearn.datasets import load_iris
iris=load_iris()
data=iris.data
def euclidean_dist_matrix(data):
    samples=len(data)
    dist_matrix=np.zeros((samples,samples))
    for i in range(samples):
        for j in range(samples):
            dist_matrix[i,j]=euclidean(data[i],data[j])
    return dist_matrix
dist_matrix=euclidean_dist_matrix(data)
print("Euclidean Distance Matrix:")
print(dist_matrix)
```

3. Import the Iris dataset. Prepare a dataset considering samples belong to any two
output classes. Draw the scatter plot for all the samples in the new dataset considering
any two input attributes. Examine the scatter plot to find the equation of a line that
can separate sample of two classes.

```python
from sklearn.datasets import load_iris
iris=load_iris()
data=iris.data
target=iris.target

index1=0
index2=1
sample1=data[target==index1]
sample2=data[target==index2]
attribute1=0
attribute2=1
plt.figure(figsize=(8,6))
plt.scatter(sample1[:,attribute1],sample1[:,attribute2],label=f"Class {index1}",color='blue')
plt.scatter(sample2[:,attribute1],sample2[:,attribute2],label=f"Class {index2}",color='red')
plt.xlabel(f"Attribute {attribute1}")
plt.ylabel(f"Attribute {attribute2}")
plt.title("Scatter Plot of Samples fron Two Output Classes: ")
plt.legend()
plt.grid(True)
plt.show()
```
